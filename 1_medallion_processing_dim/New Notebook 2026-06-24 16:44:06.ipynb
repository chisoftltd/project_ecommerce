```python
%pip install geopandas

import geopandas as gpd
import matplotlib.pyplot as plt

# 1. Load the UK regions GeoJSON file
# You can find regional and local boundary files on the UK-GeoJSON repository
# (e.g., https://martinjc.github.io/UK-GeoJSON/)
geojson_url = "https://githubusercontent.com"
uk_map = gpd.read_file(geojson_url)

# 2. Plot the regions map
fig, ax = plt.subplots(figsize=(10, 12))

# Plot the map, dividing regions by colour (using the 'name' column provided in the GeoJSON)
uk_map.plot(
    column="name",
    ax=ax,
    cmap="Set3",
    edgecolor="black",
    linewidth=0.5,
    legend=True,
    legend_kwds={
        "title": "UK Regions",
        "loc": "upper left",
        "bbox_to_anchor": (1, 1),
    },
)

# 3. Clean up the chart
ax.set_title("Map of the United Kingdom Divided by Region", fontsize=16)
ax.set_axis_off()  # Hides latitude and longitude axis lines
plt.tight_layout()

# 4. Show the map
plt.show()
```
